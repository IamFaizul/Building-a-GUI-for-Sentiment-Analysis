# Merge

```python
# There are total 4 types of join - inner,outer(full),left and right join
```

```python
# inner join (It joins multiple tables based on the common column but it only keeps the records of those which
# are present in both the table)


# syntax - left_table.merge(right_table,how='type_of_join',on='common_column')

df1 = students.merge(regs,how='inner',on='student_id')
df1

# in regs df we had also students id of 42,50,38 but in students df we just had student id upto 25. So when we merged
# both datasets we are seeing that we have record of 25 students (since they are common in both dataset)
```

**Left join** - Left join basically priotise the left table most. When we are joining two tables and performing left join, the data of the left table will always keep intact regardless of common in both datasets. So in merged dataset we will see every record of left join.

Here we are joining courses and regs dataset based on course_id. So we can see that no students has ever bought the courses of numpy and C++ since the corresponding course id of these courses are absent in regs dataset.But if use left join and keep courses dataset as left table we are gonna get the data of numpy and C++ also since they are data of the left table even though they don't have any information in the regs dataset. But if we used inner join here the record of numpy and C++ won't be included since they are not common in both table.

```python
courses.merge(regs,how='left',on='course_id')
```

```python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students
```

**Right join:** It priotises the right table.All data of the right table will be included in the merged dataset regardless of that is common in both table or not

```python
students.merge(regs,how='right',on='student_id')
```

```python
# Outer join - data of both dataset (common + not common both)

students.merge(regs,how='outer',on='student_id')
```

```python
# find the total revenue generated

rev = regs.merge(courses,how='inner',on='course_id')
rev['price'].sum()
```

```python
# find month by month revenue
nov_rev = nov.merge(courses,how='inner',on='course_id')['price'].sum()
dec_rev = dec.merge(courses,how='inner',on='course_id')['price'].sum()
print(f'Revenue of November month is {nov_rev}')
print(f'Revenue of December month is {dec_rev}')



# alternative solution

temp_df = pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
merdat = courses.merge(temp_df,how='inner',on='course_id')
merdat.groupby('level_0')['price'].sum()
```

```python
# When the datasets have different column names- left_on , right_on

import pandas as pd

# Sample DataFrames
regs = pd.DataFrame({
    'reg_id': [1, 2, 3],
    'student_identifier': [101, 102, 103],
    'course': ['Math', 'Science', 'History']
})

students = pd.DataFrame({
    'student_id': [101, 102, 103],
    'name': ['Alice', 'Bob', 'Charlie']
})

# Merging DataFrames
merged_df = regs.merge(students, left_on='student_identifier', right_on='student_id')

# Grouping, counting, sorting, and selecting top 3 rows
result = merged_df.groupby(['student_id', 'name']).size().reset_index(name='count').sort_values(by='count', ascending=False).head(3)

print(result)
```
